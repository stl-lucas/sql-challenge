```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:*********@localhost/employee_db')
connection = engine.connect()
```

```python
salary_df = pd.read_sql_query("""SELECT * FROM salaries""", connection)
title_salary_df = pd.read_sql_query("""SELECT s.emp_no, s.salary, t.title FROM salaries s JOIN titles t ON s.emp_no = t.emp_no;""", connection)
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
plt.hist(salary_df['salary'])
plt.title("Salary Ranges for Employees")
plt.xlabel("Salary")
plt.ylabel("# of Employees")
```

```python
# Create a bar chart of average salary by title.
title_gb = title_salary_df.groupby(['title'])
avg_salary = round(title_gb['salary'].mean(), 2)
plt.figure(figsize=(20, 4))
plt.bar(avg_salary.index, avg_salary)
plt.title("Avg Salary by Title")
plt.xlabel("Title")
plt.ylabel("Avg Salary")
```
